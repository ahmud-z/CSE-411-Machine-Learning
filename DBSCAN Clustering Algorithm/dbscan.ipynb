```python
import pandas as pd
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
df.head(2)
```

```python
eps_value = 1.9
min_points = 4
```

```python
dbscan_model = DBSCAN(eps=eps_value, min_samples=min_points)
dbscan_model
```

```python
predicted_cluster = dbscan_model.fit_predict(df[['X', 'Y']])
predicted_cluster
```

```python
df['cluster'] = predicted_cluster
df.head(2)
```

```python
cluster1 = df[df['cluster']==0]
cluster2 = df[df['cluster']==1]
neg_cluster = df[df['cluster']==-1]
```

```python
plt.scatter(cluster1['X'],cluster1['Y'], color = 'blue')
plt.scatter(cluster2['X'],cluster2['Y'], color='black')
plt.scatter(neg_cluster['X'],neg_cluster['Y'], color='red')
plt.show()
```
